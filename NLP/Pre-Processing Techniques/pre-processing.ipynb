```python
corpus ="""Tokenization is the process of breaking a text into meaningful units, often called tokens. 
        These tokens could be words, phrases, or even individual characters, depending on the approach."""
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
```

```python
# Word Tokenizer
tokens = word_tokenize(corpus)

print(tokens)
```

```python
## Sentence Tokenizer

sent_token = sent_tokenize(corpus)

print(sent_token)
```

```python
# Subword Tokenizer
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
tokens = tokenizer.tokenize("internationalization")

print(tokens)
```

## Stemming
Stemming is an NLP text preprocessing technique that reduces words to their root form by stripping off prefixes and suffixes.

```python
words = [
    "running", "jumps", "easily", "happily", "studies", "cats",
    "went", "better", 
    "arguing", "agreement", "connection", "flying",
    "programming", "tokenizer", "automated"
]
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word + "------>"+ stemming.stem(word))
```

## Lemmatization
Different words may have the same meaning.
Lemmatization converts them into the dictionary (base) form, called a lemma.

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("running", pos="v"))
```

## Stop Words Removal


```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

text = ["I", "love", "learning", "NLP"]

filtered = [word for word in text if word.lower() not in stop_words]

print(filtered)
```

## Part-of-Speech (POS) Tagging
POS tagging labels each word with its grammatical role.

Think of assigning jobs in a company:

Noun → Person/Thing

Verb → Action

Adjective → Description

Adverb → Describes an action

```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')

text = nltk.word_tokenize("The smart boy runs fast")
print(nltk.pos_tag(text))
```

## Named Entity Recognition (NER)
NER finds and classifies important real-world entities in text.

Examples include:

Person

Organization

Location

Date

Money

Product

```python
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp("Sundar Pichai visited India on Monday.")

for ent in doc.ents:
    print(ent.text, ent.label_)
```
