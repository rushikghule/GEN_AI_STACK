### Deep Agents
Deep Agents comes with the following built-in capabilities:
1. Take actions in an environment: Take actions via tools, read and write files, execute code
2. Connect to your data: Load memories, skills, and domain knowledge at the right moment
3. Manage growing context: Summarize history and offload large results across long runs
4. Parallelize tasks: Delegate to general or specialized subagents running in isolated context windows
5. Stay in the loop: Pause for human approval at critical decision points
6. Improve over time: Update memory, skills, and prompts based on real usage
