## Two Sum

```python
def two_sum(nums,target):
    """ using Hashmap """
    key_value_pair = {}
    for i,num in enumerate(nums):
        complement = target - num
        if complement in key_value_pair:
            return [key_value_pair[complement],i]

        key_value_pair[num]=i
    return []
```

```python
two_sum([1,2,3],5)
```

```python
def marks_grade_calc(marks):
    grade_map = {
        (0, 34): "You are failed",
        (35, 50): "You are in D grade",
        (51, 75): "You are in C grade",
        (76, 99): "You are in B grade",
        (100, 100): "You are in A grade"
    }

    if marks < 0 or marks > 100:
        return "Invalid marks"

    for (low, high), grade in grade_map.items():
        if low <= marks <= high:
            return grade


try:
    marks = int(input("Enter marks: "))
    print(marks)
    print(marks_grade_calc(marks))
except ValueError:
    print("Invalid input! Please enter numeric marks only.")
```

```python
def grade_mark(mark):
    grade_mapping = {
        (0, 34): "Failed",
        (35, 50): "in Grade D",
        (51, 75): "in Grade C",
        (76, 99): "in Grade B",
        (100, 100): "in Grade A"
    }

    if mark < 0 or mark > 100:
        return "Invalid marks"

    for (low, high), grade in grade_mapping.items():
        if low <= mark <= high:
            return f"You are {grade} because you got {mark}"


try:
    user_input = int(input("Enter marks: "))
    print(grade_mark(user_input))
except ValueError:
    print("Invalid input: Please enter numeric marks only.")
```

```python
class InvalidMarksError(Exception):
    pass
```

```python
def grade_mark(mark):
    if mark < 0 or mark > 100:
        raise InvalidMarksError(f"Marks out of range: {mark}")

    if mark < 35:
        return "Failed"
    elif mark <= 50:
        return "Grade D"
    elif mark <= 75:
        return "Grade C"
    elif mark <= 99:
        return "Grade B"
    else:
        return "Grade A"
```

```python
try:
    try:
        user_input = int(input("Enter marks: "))
    except ValueError as e:
        raise InvalidMarksError("Marks must be a number") from e

    print(grade_mark(user_input))

except InvalidMarksError as e:
    print("Error:", e)
```

### Switch Case Statements

```python

day = int(input("Enter a number (1-7): "))

# Match-case (Python 3.10+ feature) to act like a switch
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        # Default case if no match
        print("Invalid")
```

```python

# Define two constant integers
x = 10
y = 10

# Evaluate the sum of x and y using match-case (Python 3.10+)
match x + y:
    case 15:  # If the sum equals 15
        print("Result is 15.")
    case 20:  # If the sum equals 20
        print("Result is 20.")
    case _:   # Default case (no match)
        print("No match found.")

```

```python

# Define a character variable representing a grade
grade = 'B'

# Evaluate grade using match-case (Python 3.10+)
match grade:
    case 'A':  # If grade is 'A'
        print("Excellent!")
    case 'B':  # If grade is 'B'
        print("Good!")
    case _:    # Default case (no match)
        print("Not specified.")
```
