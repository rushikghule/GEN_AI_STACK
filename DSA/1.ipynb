## Two Sum

```python
def two_sum(nums,target):
    """ using Hashmap """
    key_value_pair = {}
    for i,num in enumerate(nums):
        complement = target - num
        if complement in key_value_pair:
            return [key_value_pair[complement],i]

        key_value_pair[num]=i
    return []
```

```python
two_sum([1,2,3],5)
```
